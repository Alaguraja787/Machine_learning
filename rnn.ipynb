```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras import datasets, layers, models
```

```python
data = keras.datasets.mnist
```

```python
(train_image, train_label) , (test_image, test_label) = data.load_data()
```

```python
train_image , test_image = train_image/255 , test_image/255
```

```python
train_image.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(train_image[59999])
```

```python
train_image = train_image.reshape(60000, 28,28)
test_image = test_image.reshape(10000, 28,28)
```

```python
from tensorflow.keras.utils import to_categorical
train_label = to_categorical(train_label)
test_label = to_categorical(test_label)
```

```python
from tensorflow.keras.layers import Input , LSTM, Dense
from tensorflow.keras.models import Model
input_shape = (28,28)
input = Input(shape=input_shape)
rnn_layer = LSTM(128)(input)
output = Dense(10, activation='softmax')(rnn_layer)
model = Model(inputs = input, outputs=output)
```

```python
model.compile(optimizer= 'adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(train_image,train_label, validation_split=0.1, epochs=10)
```
