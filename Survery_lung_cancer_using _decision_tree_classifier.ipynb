##**Data preprocessing**
  1.checking outliers

  2.Handling outliers
  
  3.label Encoding


##**checking the outliers**

```python
for i in df.columns:
  if df[i].dtypes !="object":
    sns.boxplot(df[i])
    plt.title("Checking for the outliers")
    plt.xlabel(i)
    plt.ylabel("Count")
    plt.show()
```

**Observation:** From the above observation only in age column outliers are present

##**Handling the outliers**

```python
#The information in the age column is important so, i am going to transform the values in between a range to reduce the gap
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method = "box-cox")
df['AGE'] = pt.fit_transform(df[["AGE"]])
```

```python
df
```

```python
sns.boxplot(df["AGE"])
plt.title("Checking for the outliers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

##**Encoding process**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in df.columns:
  if df[i].dtypes == "object":
    df[i] = le.fit_transform(df[i])
```

```python
x = df.iloc[:,:-1]
y = df[["LUNG_CANCER"]]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dc = DecisionTreeClassifier()
dc.fit(x_train, y_train)
y_pred = dc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("Performance of my model: ", accuracy_score(y_test, y_pred)*100)
```

```python
from sklearn import tree
plt.figure(figsize=(12,20))
tree.plot_tree(dc,max_depth=15)
plt.show()
```
